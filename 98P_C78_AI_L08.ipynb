```python
# =====================================
# Lesson 8: AI Career Counselor Bot
# =====================================

# üß† AI Concept:
# Simulates how AI chatbots use structured rules and user inputs
# to suggest personalized career paths.

# =========================
# Section 1: Welcome & Instructions
# =========================

print("üíº Welcome to the AI Career Counselor Bot!")
print("Answer a few questions, and I'll suggest a career path for you.\n")

# =========================
# Section 2: Main Interaction Loop
# =========================
while True:
    # ‚Äî Recap: input(), .strip(), .lower()
    subject     = input("1Ô∏è‚É£ What's your favorite school subject? ").strip().lower()
    hobby       = input("2Ô∏è‚É£ What's your favorite hobby? ").strip().lower()
    personality = input("3Ô∏è‚É£ Describe your personality (e.g., creative, analytical): ").strip().lower()

    # === Additional Activity: Ask for Dream Job ===
    # Recap: input handling + string methods
    dream_job = input("4Ô∏è‚É£ What is your dream job? ").strip().lower()

    # =========================
    # Section 3: Base Suggestions by Subject
    # =========================
    subject_suggestions = {
        "math":        "Data Scientist or Actuary",
        "science":     "Research Scientist or Engineer",
        "history":     "Archivist or Museum Curator",
        "art":         "Graphic Designer or Animator",
        "english":     "Content Writer or Editor",
        "computer":    "Software Developer or Systems Analyst"
    }
    base = subject_suggestions.get(
        subject,
        "Explore interdisciplinary fields in tech and creativity"
    )

    # =========================
    # Section 4: Adjust by Hobby
    # =========================
    if "music" in hobby:
        base += " with a side in Audio Engineering"
    elif "gaming" in hobby:
        base += " with a focus on Game Development"
    elif "sports" in hobby:
        base += " with applications in Sports Analytics"

    # =========================
    # Section 5: Adjust by Personality
    # =========================
    if "analytical" in personality:
        base += " ‚Äî your analytical mind suits data-driven roles"
    elif "creative" in personality:
        base += " ‚Äî your creativity shines in design-oriented careers"
    elif "outgoing" in personality:
        base += " ‚Äî consider roles in UX or team leadership"

    # =========================
    # === Additional Activity Logic ===
    # Modify suggestion based on the new ‚Äòdream_job‚Äô parameter
    # This reinforces if-elif-else + string checks exactly within ILO scope.
    if "engineer" in dream_job:
        base += " ‚Äî since you dream of engineering, consider Robotics Engineering"
    elif "artist" in dream_job:
        base += " ‚Äî with your artist ambitions, Digital Art could be perfect"
    elif "scientist" in dream_job:
        base += " ‚Äî a career in Environmental or Data Science fits well"
    # Else: leave base unchanged if dream_job doesn‚Äôt match any keyword

    # =========================
    # Section 6: Final Output
    # =========================
    print("\nü§ñ Career Suggestion:")
    print(f"‚Ä¢ {base}\n")

    # =========================
    # Section 7: Repeat or Exit
    # =========================
    again = input("üîÑ Would you like another suggestion? (yes/no): ").strip().lower()
    if again != "yes":
        print("\nüôè Thank you for using the AI Career Counselor Bot!")
        break

# =========================
# Recap of Concepts Used
# =========================
# - input(), .strip(), .lower() for clean user inputs
# - while loop to allow repeated interactions
# - dictionaries for base suggestions by subject
# - if-elif-else logic for hobby, personality, and dream-job tweaks
# - string concatenation & f-strings for clear, formatted output
```
