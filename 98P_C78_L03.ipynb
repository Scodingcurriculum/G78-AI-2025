```python
# =====================================
# Lesson 3: AI Age Predictor
# =====================================

# 🧠 AI Concept:
# Simulates how AI systems predict future outcomes (like age-based health tracking)
# and personalize advice based on user characteristics.

# =========================
# Section 1: Introduction
# =========================

print("=" * 50)
print("🔮 Welcome to the AI Age Predictor!")
print("This AI tool will calculate the year you'll turn 100 and give advice.")
print("=" * 50 + "\n")

# =========================
# Section 2: Collect User Info
# =========================

name = input("👤 Enter your name: ").strip().title()
# Store user's name, capitalize each word

age = int(input("🎂 Enter your age: ").strip())
# Convert input string to integer for arithmetic

current_year = 2025  # Reference year for all calculations

# =========================
# Section 3: Calculate Year to Turn 100
# =========================

years_left = 100 - age
# Compute years remaining until 100

year_100 = current_year + years_left
# Determine calendar year when user will turn 100

# =========================
# Section 4: Provide Health-Based Advice
# =========================

print(f"\nHi {name}, you will turn 100 in the year {year_100}.")
# Dynamically formatted string to show result

if age < 13:
    print("💡 Advice: You're young! Keep learning and playing.")
elif age < 40:
    print("💡 Advice: Stay active and balance your work-life!")
else:
    print("💡 Advice: Take care of your health with regular checkups.")

# =========================
# Additional Activity: Diet & Lifetime Stats
# =========================

# 1) Diet suggestion based on age group
print("\n🍎 Diet Suggestion Based on Age Group:")
if age < 13:
    print("- Drink milk daily and eat plenty of fruits for growth.")
elif age < 40:
    print("- Include lean proteins, vegetables, and 8 glasses of water daily.")
else:
    print("- Opt for a low-fat diet rich in fiber and maintain regular hydration.")

# 2) Lifetime calculation: months and days lived
print("\n📈 Lifetime Statistics:")
months_lived = age * 12
# Calculate months using multiplication (age × 12)
days_lived   = age * 365
# Approximate days using multiplication (ignores leap years)

print(f"You have lived for approximately {months_lived} months.")
# Use f-string to insert variable
print(f"You have lived for approximately {days_lived} days.")

# 3) Conditional messaging for under-18s vs. adults
print("\n🔍 Reflection:")
if age < 18:
    # Encourage young users
    print("⏳ You're under 18—enjoy every moment of learning and discovery!")
else:
    # Encourage adults
    print("⏱️ As an adult, every month and day contributes to your goals—make them count!")

# =========================
# Final Recap
# =========================

print("\n✅ You've practiced:")
print("- input() and int() for data collection and conversion")
print("- Arithmetic operators: subtraction, multiplication")
print("- if-elif-else logic for personalized messages")
print("- f-strings for dynamic, formatted output")
print("=" * 50)
```
