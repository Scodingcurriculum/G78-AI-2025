```python
# =====================================
# Lesson 6: AI History Quiz Bot
# =====================================

# üß† AI Concept:
# Simulates how AI learning assistants evaluate users using if-else logic
# and provide dynamic feedback based on performance.

# =========================
# Section 1: Introduction
# =========================

print("üß† Welcome to the AI History Quiz Bot!")
print("Answer the following questions to test your AI knowledge.\n")

# =========================
# Section 2: Initialize Counters & Storage
# =========================

score = 0                       # Counter for correct answers
wrong_answers = {}              # Store wrong Q&A for review

# =========================
# Section 3: Questions Using If-Else
# =========================

# Q1
q1 = "1. Who proposed the Turing Test?"
ans1 = input(f"{q1} ").strip().lower()
if "turing" in ans1:
    print("‚úÖ Correct!")
    score += 1
else:
    print("‚ùå Incorrect. The correct answer is Alan Turing.")
    wrong_answers[q1] = "Alan Turing"

# Q2
q2 = "2. What was the name of the first chatbot created in 1966?"
ans2 = input(f"{q2} ").strip().lower()
if "eliza" in ans2:
    print("‚úÖ Correct!")
    score += 1
else:
    print("‚ùå Incorrect. The correct answer is ELIZA.")
    wrong_answers[q2] = "ELIZA"

# Q3
q3 = "3. Which AI system won Jeopardy in 2011?"
ans3 = input(f"{q3} ").strip().lower()
if "watson" in ans3:
    print("‚úÖ Correct!")
    score += 1
else:
    print("‚ùå Incorrect. The correct answer is IBM Watson.")
    wrong_answers[q3] = "IBM Watson"

# =========================
# Section 4: Final Score & Feedback
# =========================

print(f"\nüìù You scored {score} out of 3.\n")
if score == 3:
    print("üåü AI Expert! You're outstanding!")
elif score == 2:
    print("üí° Almost there! A bit more study and you'll ace it!")
else:
    print("üìò Keep exploring the world of AI‚Äîgreat things take time!")

# =========================
# Section 5: Additional Activity
#   Track & Display Wrong Answers
# =========================

print("\nüîç Review of Incorrect Answers:")
if wrong_answers:
    for question, correct in wrong_answers.items():
        print(f" ‚Ä¢ {question} ‚Üí Correct Answer: {correct}")
else:
    print("üéâ No mistakes! You got every question right!")

# =========================
# Section 6: Extended Additional Activity
#   Retry Wrong Questions
# =========================

if wrong_answers:
    retry = input("\nüîÑ Would you like to retry the questions you got wrong? (yes/no): ").strip().lower()
    if retry == "yes":
        corrected = {}  # Track those corrected on retry
        for question, correct in wrong_answers.items():
            # Prompt retry
            new_ans = input(f"\nRetry: {question} ").strip().lower()
            if correct.lower() in new_ans:
                print("‚úÖ Now correct!")
                corrected[question] = True
            else:
                print(f"‚ùå Still incorrect. The answer remains: {correct}")
                corrected[question] = False

        # Summary of retry results
        print("\nüìà Retry Results:")
        for question, was_correct in corrected.items():
            status = "Corrected" if was_correct else "Not Corrected"
            print(f" ‚Ä¢ {question} ‚Üí {status}")

# =========================
# Section 7: Recap of Concepts Used
# =========================

print("\n‚úÖ You practiced:")
print("- if-else statements to check answers")
print("- Counters to track correct responses")
print("- Dictionary to store wrong questions and correct answers")
print("- Looping to iterate over dictionary items")
print("- Nested logic for retry mechanism")
print("- Formatted printing with f-strings")
print("=" * 50)
```
