```python
# =====================================
# Lesson 2: My AI Profile
# =====================================

# ðŸ§  AI Concept:
# This program simulates how an AI assistant collects, stores, and uses
# user preferences to customize the interaction. We use variables,
# input(), and f-strings to build a dynamic, formatted profile.

# =========================
# Section 1: Welcome Message (Lesson 1 Recap)
# =========================

print("=" * 50)
print("ðŸ¤– Welcome to the AI Profile Generator!")
print("This AI will collect your preferences and display a smart profile.")
print("=" * 50 + "\n")

# =========================
# Section 2: Collecting Basic Info using input() and variables
# =========================

name    = input("ðŸ‘¤ What is your name? ").strip().title()
age     = input("ðŸŽ‚ How old are you? ").strip()
fav_ai  = input("ðŸ’¬ Which AI assistant do you like the most (e.g., Alexa, Siri, ChatGPT)? ").strip().title()
ai_use  = input("ðŸ§  What is your favorite AI use case (e.g., translation, games, health tracking)? ").strip().lower()

# =========================
# Section 3: Display Profile Using f-strings
# =========================

print("\nðŸ“„ Your AI Profile Summary:")
print(f"Hi {name}! ðŸ‘‹ You are {age} years old.")
print(f"You like using {fav_ai} and love its ability to help with {ai_use}.")
print("Thatâ€™s awesome!\n")

# =========================
# Section 4: Lesson 1 Concept Reuse
# =========================

print("ðŸ§¾ AI Data Log (Formatted View):")
print("Name:\t\t", name)
print("Age:\t\t", age)
print("AI Tool:\t", fav_ai)
print("Use Case:\t", ai_use, "\n")

# =========================
# Section 5: Data Type Example & Categorization
# =========================

age_num = int(age)  # Convert age to integer for logic
if age_num < 13:
    category = "Young Explorer"
else:
    category = "Teen Technologist"

print(f"ðŸ§¬ Based on your age, you are categorized as a: {category}")

# ===========================================
# Extended Additional Activity: Deepen the Profile
# ===========================================

country         = input("\nðŸŒ Which country are you from? ").strip().title()
city            = input("ðŸ™ï¸ Which city do you live in? ").strip().title()
profession      = input("ðŸŽ¯ What do you want to be when you grow up? ").strip().title()
favorite_feature= input("â­ What AI feature impressed you the most? ").strip()
language        = input("ðŸ’» Which programming language would you like to learn next? ").strip().title()

# Build a multi-line extended profile summary using an f-string
extended_profile = (
    f"\nðŸ“ Extended AI Profile for {name}:\n"
    f"{'-'*40}\n"
    f"Name:                   {name}\n"
    f"Age:                    {age_num} years\n"
    f"Category:               {category}\n"
    f"Country:                {country}\n"
    f"City:                   {city}\n"
    f"Favorite AI Assistant:  {fav_ai}\n"
    f"Favorite Use Case:      {ai_use}\n"
    f"Desired Profession:     {profession}\n"
    f"Top AI Feature:         {favorite_feature}\n"
    f"Next Language to Learn: {language}\n"
    f"{'-'*40}"
)

# Print the extended profile
print(extended_profile)

# ===========================================
# Final Recap of Concepts Used
# ===========================================

print("\n" + "=" * 50)
print("âœ… Concepts Reinforced:")
print("- input() for dynamic user data")
print("- Variable storage: name, age, preferences, country, city, profession, feature, language")
print("- Type conversion using int()")
print("- f-string formatting for multi-line, aligned output")
print("- print() formatting with separators and alignment")
print("=" * 50)
```
